```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
from bc_augmentation import *
from bc_preprocessing import *
import cv2
data_path = 'CAR_data/data/'
img_path = ''
csv_path = 'driving_log.csv'

csv_data = pd.read_csv(data_path+csv_path,usecols=["center","left","right","steering"])
sel_idx = np.random.randint(len(csv_data))

line_data = csv_data.iloc[[sel_idx]].reset_index()

left_image_file =  data_path+img_path+line_data['left'][0].strip()
right_image_file =  data_path+img_path+line_data['right'][0].strip()
center_image_file =  data_path+img_path+line_data['center'][0].strip()
sa = float(line_data['steering'])

cimg = mpimg.imread(center_image_file)
bimg = augment_brightness(cimg)
tr_img,s_a = augment_by_translation(cimg,sa,100,40)
fl_img = cv2.flip(cimg,flipCode=1)
rs_img = crop_resize(cimg,66,220)

f, (ax1, ax2,ax3,ax4,ax5) = plt.subplots(5, 1)
ax1.imshow(cimg)
ax1.set_title('Original Image')
ax1.axis('off')

ax2.imshow(tr_img)
ax2.set_title('Translated Image')
ax2.axis('off')

ax3.imshow(bimg)
ax3.set_title('Brightness Augmentation')
ax3.axis('off')

ax4.imshow(fl_img)
ax4.set_title('Flipped Image')
ax4.axis('off')


ax4.imshow(rs_img)
ax4.set_title('Resized Image')
ax4.axis('off')


```
